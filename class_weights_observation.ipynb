## metric

```python
# ture positive
def get_TP(y_true, y_pred):
    return sum((y_true==1)&(y_pred==1))

def get_FP(y_true, y_pred):
    return sum((y_true==0)&(y_pred==1))

def get_FN(y_true, y_pred):
    return sum((y_true==1)&(y_pred==0))

def get_TN(y_true, y_pred):
    return sum((y_true==0)&(y_pred==0))
```

```python
def confusion_matrix(y_true, y_pred):
    TP = get_TP(y_true, y_pred)
    FN = get_FN(y_true, y_pred)
    FP = get_FP(y_true, y_pred)
    TN = get_TN(y_true, y_pred)
    return np.array([[TN, FP],[FN, TP]])
```

```python
def accuracy(y_true, y_pred):
    TP, TN = get_TP(y_true, y_pred), get_TN(y_true, y_pred)
    return (TP + TN) / len(y_true)

def recall(y_true, y_pred):
    TP = get_TP(y_true, y_pred)
    total_pos = sum(y_true)
    return TP / total_pos

def precision(y_true, y_pred):
    TP = get_TP(y_true, y_pred)
    total_pos_pred = sum(y_pred)
    return TP / total_pos_pred
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
X, y = data.drop(columns=["Class", "Time"]), data.Class

lb, hb = 0.05, 0.95
n_points = 10
class_weights = [{0:(hb - lb)/n_points * i + lb, 1:1} for i in range(n_points + 1)]
print(class_weights)

gs = GridSearchCV(RandomForestClassifier(n_estimators=250, max_depth=3), verbose=2, return_train_score=True, scoring="f1", param_grid={"class_weight":class_weights} , cv=5, n_jobs=-1)
gs.fit(X, y)
```

```python
cls_weights = [cand["class_weight"] for cand in gs.cv_results_['params']]
test_scores = np.mean(np.array([gs.cv_results_['split{}_test_score'.format(i)] for i in range(5)]), axis=0)
for maj_weight, test_f1 in zip(cls_weights, test_scores):
    print("Weights: {:.4f} (Majority): {:.4f} (Minority), f1 score: {:.4f}".format(maj_weight[0], maj_weight[1], test_f1))
```

```python
plt.figure(figsize=(14,10))
plt.plot([cw[0] for cw in cls_weights], test_scores, 'r--')
plt.title('F1 score by majority class weight (1.0 for minority class)') 
```
